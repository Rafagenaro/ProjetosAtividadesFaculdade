```python
df[df.Diagnosis == 'M']['radius'].plot.kde(bw_method=0.3, color='red', label = 'Maligno')
df[df.Diagnosis == 'B']['radius'].plot.kde(bw_method=0.3, color='blue', label = 'Benigno')
plt.legend()
plt.show()
```

```python
df[df.Diagnosis == 'M']['concave_points'].plot.kde(bw_method=0.5, color='red', label = 'Maligno')
df[df.Diagnosis == 'B']['concave_points'].plot.kde(bw_method=0.5, color='blue', label = 'Benigno')
plt.legend()
plt.show()
```

```python
df[df.Diagnosis == 'M']['compactness'].plot.kde(bw_method=0.3, color='red', label = 'Maligno')
df[df.Diagnosis == 'B']['compactness'].plot.kde(bw_method=0.3, color='blue', label = 'Benigno')
plt.legend()
plt.show()
```

```python
df[df.Diagnosis == 'M']['compactness'].plot.kde(bw_method=0.5, color='red', label = 'Maligno_compac')
df[df.Diagnosis == 'B']['compactness'].plot.kde(bw_method=0.5, color='blue', label = 'Benigno_compac')
df[df.Diagnosis == 'M']['concave_points'].plot.kde(bw_method=0.5, color='green', label = 'Maligno_concave')
df[df.Diagnosis == 'B']['concave_points'].plot.kde(bw_method=0.5, color='orange', label = 'Benigno_concave')
df[df.Diagnosis == 'M']['radius'].plot.kde(bw_method=0.5, color='yellow', label = 'Maligno_radius')
df[df.Diagnosis == 'B']['radius'].plot.kde(bw_method=0.5, color='pink', label = 'Benigno_radius')
plt.legend()
plt.show()
```

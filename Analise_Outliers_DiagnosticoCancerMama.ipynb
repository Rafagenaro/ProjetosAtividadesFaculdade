```python
for feature in df.drop('Diagnosis', axis=1).columns:
  df[df.Diagnosis == 'M'][feature].plot.kde(bw_method= 0.2, color = 'red', label = 'Maligno')
  df[df.Diagnosis == 'B'][feature].plot.kde(bw_method= 0.2, color = 'blue', label = 'Benigno')
  plt.title(f'Distribuição de {feature}')
  plt.legend(['M','B'])
  plt.show()
```

```python
df.boxplot(column = 'radius')
```

```python
for feature in df.drop('Diagnosis', axis=1).columns:

  fig, ax = plt.subplots()

  df[df.Diagnosis == 'M'].boxplot(column = feature, ax = ax, positions=[0], widths = 0.6, patch_artist = True, boxprops=dict(facecolor = 'red'))
  df[df.Diagnosis == 'B'].boxplot(column = feature, ax = ax, positions=[1], widths = 0.6, patch_artist = True, boxprops=dict(facecolor = 'blue'))

  ax.set_xticks([0,1])
  ax.set_xticklabels(['Maligno', 'Benigno'])
  ax.set_title(f'Boxplot de {feature}')

  plt.show()
```

```python
Q3B = df[df.Diagnosis == 'B'].radius.quantile(0.75)
M = df[df.radius > Q3B].Diagnosis.value_counts()[0]
B = df[df.radius > Q3B].Diagnosis.value_counts()[1]

print(df[df.radius > Q3B].Diagnosis.value_counts())
print(M/(M+B)*100)
```

```python
Q3B = df[df.Diagnosis == 'B'].radius.quantile(0.75)
Q3B2 = df[df.Diagnosis == 'B'].concave_points.quantile(0.75)
M = df[(df.radius > Q3B) & (df.concave_points > Q3B2)].Diagnosis.value_counts()[0]
B = df[(df.radius > Q3B) & (df.concave_points > Q3B2)].Diagnosis.value_counts()[1]

print(df[(df.radius > Q3B) & (df.concave_points > Q3B2)].Diagnosis.value_counts())
print(M/(M+B)*100)
```

```python
Q3B = df[df.Diagnosis == 'B'].radius.quantile(0.75)
Q3B2 = df[df.Diagnosis == 'B'].concave_points.quantile(0.75)
Q3B3 = df[df.Diagnosis == 'B'].compactness.quantile(0.75)

M = df[(df.radius > Q3B) & (df.concave_points > Q3B2) & (df.compactness > Q3B3)].Diagnosis.value_counts()[0]
B = df[(df.radius > Q3B) & (df.concave_points > Q3B2) & (df.compactness > Q3B3)].Diagnosis.value_counts()[1]

print(df[(df.radius > Q3B) & (df.concave_points > Q3B2) & (df.compactness > Q3B3)].Diagnosis.value_counts())
print(M/(M+B)*100)
```
